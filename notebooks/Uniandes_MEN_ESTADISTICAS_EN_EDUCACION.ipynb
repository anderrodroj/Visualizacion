### Contexto del Dataset

Este conjunto de datos consolida estadísticas educativas de Colombia a nivel territorial. Su estructura abarca:
* **Alcance temporal:** 14 años de registros consolidados (`AÑO`).
* **Alcance geográfico:** 34 categorías territoriales (`DEPARTAMENTO`), incluyendo los departamentos del país y Bogotá D.C.
* **Métricas principales:** * **Demografía y Acceso:** Población objetivo escolar, tasas de matriculación y coberturas (neta y bruta).
    * **Infraestructura escolar:** Tamaño promedio de grupo y porcentaje de sedes conectadas a internet.
    * **Trayectoria y Desempeño:** Tasas de deserción, aprobación, reprobación y repitencia. Todas estas métricas están desglosadas por nivel educativo (Transición, Primaria, Secundaria y Educación Media).

###  Limpieza y Transformación del Dataset

Para preparar los datos antes del análisis de deserción escolar, se le aplicó el siguiente tratamiento al dataset:

* **Exploración inicial:** Revisamos la estructura general, tipos de datos, valores únicos por columna y confirmamos que no hay filas duplicadas.
* **Transformación de porcentajes (0 a 1):** Más de 30 columnas (como coberturas, tasas de deserción, aprobación, etc.) venían como texto (`object`) con el símbolo `%` y comas. Se limpiaron estos caracteres y se convirtieron a tipo `float` en una escala decimal de 0 a 1.
* **Ajuste de variables específicas:** * `TAMAÑO_PROMEDIO_DE_GRUPO`: Se pasó a numérico cambiando la coma por punto.
    * `SEDES_CONECTADAS_A_INTERNET`: Se dividió entre 100 para unificarla a la misma escala decimal (0 a 1) del resto de las tasas.
* **Exportación para Excel:** Guardamos el resultado final en `datos_educacion_limpios.csv` utilizando codificación `utf-8-sig`, separador de punto y coma (`;`) y decimales con coma (`,`) para garantizar que Excel lea perfectamente las tildes, las eñes y los números.

```python
import pandas as pd
```

```python
path = "/content/MEN_ESTADISTICAS_EN_EDUCACION_EN_PREESCOLAR,_BÁSICA_Y_MEDIA_POR_DEPARTAMENTO_20260301.csv"
data = pd.read_csv(path)
```

```python
print("--- Información General y Tipos de Datos ---")
data.info()
```

```python
print("\n--- Conteo de Valores Únicos ---")
print(data.nunique())
```

```python
print("\n--- Cantidad de Filas Duplicadas ---")
total_duplicados = data.duplicated().sum()
print(f"Total de filas exactas duplicadas: {total_duplicados}")
```

```python
print("\n--- Muestra de Valores Únicos por Columna ---")
for col in data.columns:
    valores_unicos = data[col].dropna().unique()
    print(f"[{col}] - Tipo: {data[col].dtype}")
    print(f"Muestra de únicos: {valores_unicos[:5]}\n")
```

```python
cols_porcentajes = [
    'TASA_MATRICULACIÓN_5_16', 'COBERTURA_NETA', 'COBERTURA_NETA_TRANSICIÓN',
    'COBERTURA_NETA_PRIMARIA', 'COBERTURA_NETA_SECUNDARIA', 'COBERTURA_NETA_MEDIA',
    'COBERTURA_BRUTA', 'COBERTURA_BRUTA_TRANSICIÓN', 'COBERTURA_BRUTA_PRIMARIA',
    'COBERTURA_BRUTA_SECUNDARIA', 'COBERTURA_BRUTA_MEDIA', 'DESERCIÓN',
    'DESERCIÓN_TRANSICIÓN', 'DESERCIÓN_PRIMARIA', 'DESERCIÓN_SECUNDARIA',
    'DESERCIÓN_MEDIA', 'APROBACIÓN', 'APROBACIÓN_TRANSICIÓN', 'APROBACIÓN_PRIMARIA',
    'APROBACIÓN_SECUNDARIA', 'APROBACIÓN_MEDIA', 'REPROBACIÓN', 'REPROBACIÓN_TRANSICIÓN',
    'REPROBACIÓN_PRIMARIA', 'REPROBACIÓN_SECUNDARIA', 'REPROBACIÓN_MEDIA', 'REPITENCIA',
    'REPITENCIA_TRANSICIÓN', 'REPITENCIA_PRIMARIA', 'REPITENCIA_SECUNDARIA', 'REPITENCIA_MEDIA'
]

for col in cols_porcentajes:
    data[col] = data[col].astype(str).str.replace('%', '', regex=False).str.replace(',', '.', regex=False)
    data[col] = pd.to_numeric(data[col], errors='coerce') / 100

data['TAMAÑO_PROMEDIO_DE_GRUPO'] = data['TAMAÑO_PROMEDIO_DE_GRUPO'].astype(str).str.replace(',', '.', regex=False)
data['TAMAÑO_PROMEDIO_DE_GRUPO'] = pd.to_numeric(data['TAMAÑO_PROMEDIO_DE_GRUPO'], errors='coerce')

data['SEDES_CONECTADAS_A_INTERNET'] = data['SEDES_CONECTADAS_A_INTERNET'] / 100

print(data.dtypes)
```

```python
data.to_csv('datos_educacion_limpios.csv',
            index=False,
            encoding='utf-8-sig',
            sep=';',
            decimal=',')
```
